```python
import pandas as pd
# Import matplotlib for plotting and use magic command for Jupyter Notebooks
import matplotlib.pyplot as plt
%matplotlib inline
#import normalization
from sklearn import preprocessing
eu = pd.read_csv('EURUSD1440.csv')
eu.columns=['Date','Time','eu_Open','eu_High','eu_Low','eu_Close','Volume']
#eu = eu.iloc[-2000:] 
eu = eu.iloc[-8000:] 
eu = eu.reset_index(drop=True)
eu['Index']=eu.index
eu.set_index('Index', drop=True, append=False, inplace=True, verify_integrity=False)
print(eu.head())
print(eu.tail())
```

```python
new = eu['Date'].str.split('.',n=2,expand=True)
eu['Year']=new[0]
eu['Month']=new[1]
eu['Day']=new[2]
eu = eu.drop(columns=['Date'])
```

```python
eu
```

```python

eu['eu_pred_close'] = eu.eu_Close.shift(-1)
eu['eu_pred_low'] = eu.eu_Low.shift(-1)
eu['eu_pred_high'] = eu.eu_High.shift(-1)

#clean N/A values off dataset
eu = eu.dropna(how='any',axis=0)
#drop columns if needed
#eu =eu.drop(columns=['Time','eu_Open','Volume','Year','Month','Day'])
#drop additional rows if needed
#eu = eu[:-1]
eu.tail()
```

```python
features = ['eu_Close','eu_Low','eu_High','eu_Open']
target_c = 'eu_pred_close'
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators=15000)
rf.fit(eu[features],eu[target_c])
print(rf.predict([[1.16505,1.16167,1.16585,1.16303]]))
```

```python
target_l = 'eu_pred_low'
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators=15000)
rf.fit(eu[features],eu[target_l])
print(rf.predict([[1.16505,1.16167,1.16585,1.16303]]))
```

```python

target_h = 'eu_pred_high'
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators=15000)
rf.fit(eu[features],eu[target_h])
print(rf.predict([[1.16505,1.16167,1.16585,1.16303]]))
```
